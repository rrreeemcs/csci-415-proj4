```python
# Reading Algerian Forest Fires Dataset
import pandas as pd
```

```python
df = pd.read_csv('data/Algerian_forest_fires_dataset.csv', header=1)
df.head()
```

## Combining the Regions

```python
# Getting the Bejaia region indices [0:122] from df
bejaia_df = df.iloc[:122]
bejaia_df['region'] = 'Bejaia'
bejaia_df.set_index('region', inplace=True)

# Performing same process with the Sidi-Bell Abbes region [124:]
sidi_bel_abbes_df = df.iloc[124:]
sidi_bel_abbes_df['region'] = 'Sidi-Bel Abbes'
sidi_bel_abbes_df.set_index('region', inplace=True)

# Combining these parts into full df
algerian_ff = pd.concat([bejaia_df, sidi_bel_abbes_df], axis=0)
```

```python
algerian_ff.loc['Bejaia']
```

```python
algerian_ff.loc['Sidi-Bel Abbes']
```

## Formatting columns

```python
# Stripping whitespace from columns and fields
algerian_ff.columns = algerian_ff.columns.str.strip()
algerian_ff = algerian_ff.apply(lambda x: x.str.strip() if x.dtype == 'str' else x)

# Changing column names to lowercase
algerian_ff.columns = algerian_ff.columns.str.lower()
```

```python
algerian_ff.dtypes
```

```python
# Changing temperature, rh, ws, rain to float datatypes
algerian_ff = algerian_ff.astype({
    'temperature': 'float',
    'rh': 'float',
    'ws': 'float',
    'rain': 'float',
})

# Renaming above columns to match their unit of measurement
algerian_ff.rename({
    'temperature': 'temp_c',
    'rh': 'rel_humidity_percent',
    'ws': 'wind_speed_kmh',
    'rain': 'rainfall_mm',
}, axis=1, inplace=True)

algerian_ff.dtypes
```

### Dropping Rows & Columns

```python
algerian_ff.drop(['day', 'month', 'year'], axis=1, inplace=True)
```

```python
algerian_ff.isnull().sum()
```

```python
algerian_ff.dropna(axis=0, subset='classes', inplace=True)
algerian_ff.isnull().sum() > 0
```
